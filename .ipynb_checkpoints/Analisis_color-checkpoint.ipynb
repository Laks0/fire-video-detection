```python
from Color_functions import *
from skimage import io
from skimage.color import rgb2lab
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
plt.gray()
```

```python
rgb_img = io.imread("imagenes/images_prepped_train/img(1).jpg").astype(np.float32)  /255
lab_img = rgb2lab(rgb_img)
```

```python
prob_la = pd.read_json("la.json")
prob_lb = pd.read_json("lb.json")
prob_ab = pd.read_json("ab.json")
```

```python
fig, axes = plt.subplots(1,7, figsize=(20, 10))

R1, R2, R3, R4, R5 = mascaras(lab_img, prob_la, prob_lb, prob_ab, 0.005)

axes[0].imshow(rgb_img)
axes[1].imshow(R1)
axes[2].imshow(R2)
axes[3].imshow(R3)
axes[4].imshow(R4)
axes[5].imshow(R5)
axes[6].imshow(R1 & R2 & R3 & R4)
```
