```python
from Color_functions import *
from skimage import io
from skimage.color import rgb2lab
import numpy as np
import matplotlib.pyplot as plt
import json
```

```python
plt.gray()
```

```python
rgb_img = io.imread("imagenes/images_prepped_train/img(500).jpg").astype(np.float32)  /255
lab_img = rgb2lab(rgb_img)
```

```python
prob_la = np.load("prob_la.npy")
prob_lb = np.load("prob_lb.npy")
prob_ab = np.load("prob_ab.npy")
```

```python
fig, axes = plt.subplots(1,7, figsize=(20, 10))

R1, R2, R3, R4, R5 = mascaras(lab_img, prob_la, prob_lb, prob_ab, 0.0016)

axes[0].imshow(rgb_img)
axes[1].imshow(R1)
axes[2].imshow(R2)
axes[3].imshow(R3)
axes[4].imshow(R4)
axes[5].imshow(R5)
axes[6].imshow(R1 & R2 & R3 & R4 & R5)
```

```python
fig, axes = plt.subplots(1,2, figsize=(20, 10))

axes[0].imshow(rgb_img)
axes[1].imshow(rgb_img * (R1 & R2 & R3 & R4 & R5)[...,None])
```

```python
fig, axes = plt.subplots(3,3, figsize=(15,10))

R1a4 = R1&R2&R3&R4
for i in range(3):
    S1, S2, S3 = otsu_segmentos(lab_img[:,:,i])
    mejora = np.argmax([coincidencia(S1, R5), coincidencia(S2, R5), coincidencia(S3, R5)])
    mascara = np.zeros_like(R5)
    if mejora == 0:
        mascara = S1
    elif mejora == 1:
        mascara = S2
    elif mejora == 2:
        mascara = S3
    axes[i][1].imshow(rgb_img * (R5 & mascara)[...,None])
    
    mejorb = np.argmax([coincidencia(S1, R1a4), coincidencia(S2, R1a4), coincidencia(S3, R1a4)])
    if mejorb == 0:
        mascara = S1
    elif mejorb == 1:
        mascara = S2
    elif mejora == 2:
        mascara = S3
    axes[i][2].imshow(rgb_img * (R1a4 & mascara)[...,None])

    _, m = otsu_simple(lab_img[:,:,i])
    axes[i][0].imshow(rgb_img * m[...,None])

axes[0][0].set_title("Segmentación Otsu")
axes[0][1].set_title("Mejores segmentaciones con R5")
axes[0][2].set_title("Mejores segmentaciones con R1&...&R4");
```
